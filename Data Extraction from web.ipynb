```python
url="https://www.besanttechnologies.com/data-science-interview-questions-and-answers"
```

```python
from bs4 import BeautifulSoup
import requests


response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.prettify()
```

```python
for h in soup.find_all("h4",{"class":"slide_toggle clicked"}):
    print(h.get('a'))
```

```python
answer=[]
string= soup.findAll("div",attrs={"class":"slide_toggle_content"})
for div in string:
    print(div.find("p"))
    answer.append(div.find("p"))
```

```python
qsn=[]
for data in soup.find_all("h4",attrs={"class":"slide_toggle"}):
    for a in data.find_all('a'):
        print(a.text) #for g
        qsn.append(a.text)
```

```python
import pandas as pd
qbook= pd.DataFrame(list(zip(qsn, answer)), columns =['Question', 'Answer']) 
```
